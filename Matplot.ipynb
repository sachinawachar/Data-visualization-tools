```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Basic plot

```python
x=np.arange(1,9)
y=x**2
```

```python
y
```

```python

plt.plot(x,y,color='red',marker='p',linestyle='dashed',linewidth=2,markersize=12)
plt.title('line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

# Customise the height and width of figure

```python
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,'go-',linewidth=4,markersize=10)
plt.title('Line chart')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

# Default plot

```python
plt.plot(x,y)
```

# Scatter plot

```python
fig=plt.figure(figsize=(12,8))
plt.scatter(df['Horsepower'],df['Price'],color='r')
plt.title('Scatter plot')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.show()
```

# Histogram

```python
fig=plt.figure(figsize=(6,4))
plt.hist(df["Horsepower"])
plt.title("Histogram of Horsepower")
plt.show()
```
