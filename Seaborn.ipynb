```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
y=[10,23,8,18,3,25,30]
```

```python
new=sns.stripplot(x,y)
new.set(xlabel="Days",ylabel="Count")
plt.title("Strip Plot")
plt.show()
```

```python
sns.distplot([0,1,2,3,4,5])
plt.show()
```

```python
sns.distplot([0,1,2,3,4,5],hist=False)
plt.show()
```

```python
df=pd.read_csv("Iris.csv")
df
```

```python
sns.stripplot(df["Species"],df["Sepal_Length"],jitter=True)
plt.title("Strip Plot")
plt.xlabel("Species")
plt.ylabel("Sepal_Length")
plt.show()
```

```python
sns.swarmplot(df["Species"],df["Sepal_Length"])
plt.title("Strip Plot")
plt.xlabel("Species")
plt.ylabel("Sepal_Length")
plt.show()
```

```python
sns.barplot(df["Species"],df["Sepal_Length"])
plt.title("Bar Plot")
plt.xlabel("Species")
plt.ylabel("Sepal_Length")
plt.show()
```

```python
new=sns.countplot(df["Species"])
new.set(xlabel="Species",ylabel="Count")
plt.title("Count Plot of Species")
plt.show()
```

```python
new1=sns.boxplot(df["Species"],df["Sepal_Length"])
new1.set(xlabel="Species",ylabel="Count")
plt.title("Box Plot of Species vs Sepal_length")
plt.show()
```

```python
new1=sns.violinplot(df["Species"],df["Sepal_Length"])
new1.set(xlabel="Species",ylabel="Count")
plt.title("Violin Plot of Species vs Sepal_length")
plt.show()
```
